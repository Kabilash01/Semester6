```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
normal = cv2.imread('C:/E0323036-SEM6/dl_med/normal.jpeg', 0)
abnormal = cv2.imread('C:/E0323036-SEM6/dl_med/POSITIVE.jpeg', 0)
if normal is None or abnormal is None:
    print("Error: Images not found")
else:
    print("Images loaded successfully")
```

```python

abnormal = cv2.resize(abnormal, (normal.shape[1], normal.shape[0]))
difference = cv2.absdiff(normal, abnormal)
plt.figure(figsize=(12, 4))
```

```python

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(normal, cmap='gray')
plt.title('Normal X-ray')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(abnormal, cmap='gray')
plt.title('Abnormal X-ray')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(difference, cmap='viridis') 
plt.title('Difference')
plt.axis('off')
plt.tight_layout()
plt.show()

print("Comparison complete!")
```
