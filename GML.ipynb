```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4, 5, 6, 7, 8])
edges = [(1, 2), (1, 3), (2, 4), (3, 4), (4, 5), (5, 6), (6, 7), (7, 8), (5, 8)]
G.add_edges_from(edges)
print("GRAPH PROPERTIES")
print("-" * 30)
print(f"Number of Nodes: {G.number_of_nodes()}")
print(f"Number of Edges: {G.number_of_edges()}")
print(f"Nodes: {list(G.nodes())}")
print(f"Edges: {list(G.edges())}")
print(f"Degree of each node: {dict(G.degree())}")
print(f"Is Connected: {nx.is_connected(G)}")
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='lightblue',
        node_size=700, font_size=16, font_weight='bold')
plt.show()
```

```python

source = 1
target = 8
shortest_path = nx.shortest_path(G, source=source, target=target)
path_length = nx.shortest_path_length(G, source=source, target=target)

print(f"\n1. Shortest Path from Node {source} to Node {target}:")
print(f"   Path: {shortest_path}")
print(f"   Distance: {path_length}")
print(f"\n2. All Shortest Paths:")
all_paths = dict(nx.all_pairs_shortest_path(G))
for source_node in all_paths:
    for target_node in all_paths[source_node]:
        if source_node < target_node:
            print(f"   {source_node} -> {target_node}: {all_paths[source_node][target_node]}")

print(f"\n3. Distance Between All Nodes:")
all_distances = dict(nx.all_pairs_shortest_path_length(G))
for source_node in sorted(all_distances.keys()):
    for target_node in sorted(all_distances[source_node].keys()):
        if source_node < target_node:  # Avoid duplicates
            distance = all_distances[source_node][target_node]
            print(f"   Node {source_node} to Node {target_node}: {distance}")
if nx.is_connected(G):
    avg_path_length = nx.average_shortest_path_length(G)
    print(f"\n4. Average Shortest Path Length: {avg_path_length:.2f}")
else:
    print(f"\n4. Graph is not connected, cannot calculate average path length")
print(f"\n5. Neighbors of Each Node:")
for node in G.nodes():
    neighbors = list(G.neighbors(node))
    print(f"   Node {node}: {neighbors} (Total: {len(neighbors)})")
start_node = 1
print(f"\n6. Distances from Node {start_node} to All Other Nodes:")
distances_from_start = nx.single_source_shortest_path_length(G, start_node)
for node, distance in distances_from_start.items():
    print(f"   Node {start_node} to Node {node}: {distance}")
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, seed=42)

nx.draw_networkx_edges(G, pos, edge_color='gray', width=2)

path_edges = [(shortest_path[i], shortest_path[i+1]) for i in range(len(shortest_path)-1)]
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=4)
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=700)
nx.draw_networkx_nodes(G, pos, nodelist=[source], node_color='green', node_size=700)
nx.draw_networkx_nodes(G, pos, nodelist=[target], node_color='orange', node_size=700)
nx.draw_networkx_labels(G, pos, font_size=16, font_weight='bold')

plt.title(f"Shortest Path from Node {source} (Green) to Node {target} (Orange)", fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()
```
